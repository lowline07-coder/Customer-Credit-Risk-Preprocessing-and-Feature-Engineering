## Q1. What is Data Analysis?

### Definition

Data Analysis is the process of collecting, organizing, cleaning, transforming, and interpreting data to discover useful information, identify patterns, and support decision-making.

### Objectives of Data Analysis

1. Understand the structure and characteristics of data.
2. Identify patterns and trends.
3. Detect anomalies and outliers.
4. Support business decision-making.
5. Prepare data for Machine Learning models.

### Steps in Data Analysis

1. Data Collection
2. Data Cleaning
3. Data Transformation
4. Exploratory Data Analysis (EDA)
5. Data Visualization
6. Interpretation and Reporting

### Importance of Data Analysis

Data Analysis helps organizations make informed decisions, improve operational efficiency, reduce risks, and gain valuable insights from data. It is a crucial step before building Machine Learning models because the quality of analysis directly affects model performance.



## Q2. How to Plan a Data Science Project?

A Data Science Project follows a systematic workflow to solve real-world problems using data.

### Step 1: Business Understanding

Understand the problem and define project objectives.

**Example:** Predict whether a customer will default on a loan.

### Step 2: Data Collection

Gather data from multiple sources such as:

* CSV files
* Databases
* APIs
* JSON files
* Excel spreadsheets

### Step 3: Data Understanding

Analyze the dataset to identify:

* Data types
* Missing values
* Outliers
* Data distributions

### Step 4: Data Cleaning

Improve data quality by:

* Handling missing values
* Removing duplicates
* Correcting inconsistencies
* Treating outliers

### Step 5: Feature Engineering

Create meaningful features from existing variables.

**Example:**

Debt-to-Income Ratio = Loan Amount / Annual Income

Spending-To-Income Ratio = Spending / Income

### Step 6: Model Building

Train Machine Learning algorithms using prepared data.

### Step 7: Model Evaluation

Evaluate performance using metrics such as:

* Accuracy
* Precision
* Recall
* F1 Score 



### Step 8: Deployment

Deploy the model into production environments using tools like Flask, FastAPI, or Streamlit.

### Step 9: Monitoring and Maintenance

Monitor model performance and retrain when necessary.

### Conclusion

A well-planned Data Science Project improves efficiency, reduces project risks, and ensures successful implementation.

## Q4. Explain Tensors with NumPy Examples

### Definition

A Tensor is a mathematical object used to represent data in multiple dimensions. It is a generalized form of scalars, vectors, and matrices.

Tensors are widely used in:

* Machine Learning
* Deep Learning
* Computer Vision
* Natural Language Processing



### Types of Tensors

#### 0-D Tensor (Scalar)

Example:

5

A scalar contains only one value and has zero dimensions.



#### 1-D Tensor (Vector)

Example:

[1, 2, 3, 4]

A vector is a collection of values arranged in a single dimension.



#### 2-D Tensor (Matrix)

Example:

[[1, 2],
[3, 4]]

A matrix consists of rows and columns and has two dimensions.


#### 3-D Tensor

Example:

[[[1, 2],
[3, 4]],

[[5, 6],
[7, 8]]]

A 3-D Tensor consists of multiple matrices stacked together.



### Tensor Operations

#### Addition

Two tensors of the same shape can be added element-wise.

Example:

[1, 2, 3] + [4, 5, 6]

Output:

[5, 7, 9]



#### Multiplication

Element-wise multiplication:

[1, 2, 3] × [4, 5, 6]

Output:

[4, 10, 18]



#### Matrix Multiplication

Matrix multiplication combines rows and columns of matrices to produce a new matrix.



#### Transpose

The transpose of a matrix swaps rows and columns.

Example:

[[1, 2],
[3, 4]]

Transpose:

[[1, 3],
[2, 4]]



#### Reshaping

Reshaping changes the dimensions of a tensor without changing its data.

Example:

Convert a 1-D array into a 2-D matrix.



### Applications of Tensors

#### Computer Vision

Images are represented as tensors.

Example:

Height × Width × Channels

224 × 224 × 3



#### Natural Language Processing

Word embeddings and text representations are stored as tensors.
 


#### Deep Learning

Neural networks process tensors during training and prediction.



### Conclusion

Tensors are the fundamental data structures used in Machine Learning and Deep Learning. They allow efficient storage, manipulation, and processing of multidimensional data.

### PART D – OUTLIER DETECTION

Z-Score Method

```python
from scipy.stats import zscore
import numpy as np

z = np.abs(
    zscore(df["annual_income"])
)

df_z = df[z < 3]

print("Original Shape:", df.shape)

print("Shape After Z-Score:", df_z.shape)

print(
    "Rows Removed:",
    df.shape[0] - df_z.shape[0]
)

print(df_z.head())
```

IQR Method

```python
Q1 = df["annual_income"].quantile(0.25)

Q3 = df["annual_income"].quantile(0.75)

IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR

upper = Q3 + 1.5 * IQR

df_iqr = df[
    (df["annual_income"] >= lower)
    &
    (df["annual_income"] <= upper)
]
print("Original Shape:", df.shape)
print("New Shape:", df_iqr.shape)
print("Outliers Removed:", df.shape[0] - df_iqr.shape[0])
```

Percentile Method

```python
lower = df["annual_income"].quantile(0.01)

upper = df["annual_income"].quantile(0.99)

df_percentile = df[
    (df["annual_income"] >= lower)
    &
    (df["annual_income"] <= upper)
]
print("Original Shape:", df.shape)
print("New Shape:", df_percentile.shape)    
```

Winsorization

```python
from scipy.stats.mstats import winsorize

df["annual_income"] = winsorize(
    df["annual_income"],
    limits=[0.05,0.05]
)
print(df["annual_income"].describe())
```

### COLUMN TRANSFORMER

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
import pandas as pd

num_cols = [
    "annual_income",
    "loan_amount",
    "credit_score"
]

cat_cols = [
    "gender"
]

preprocessor = ColumnTransformer(
    transformers=[
        (
            "num",
            StandardScaler(),
            num_cols
        ),
        (
            "cat",
            OneHotEncoder(
                drop="first",
                sparse_output=False
            ),
            cat_cols
        )
    ]
)

transformed_data = (
    preprocessor.fit_transform(df)
)

column_names = (
    num_cols
    +
    list(
        preprocessor
        .named_transformers_["cat"]
        .get_feature_names_out(cat_cols)
    )
)

transformed_df = pd.DataFrame(
    transformed_data,
    columns=column_names
)

print("Column Transformer Applied Successfully\n")

print(transformed_df.head(10))

print("\nShape\n")

print(transformed_df.shape)
```
